```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer(as_frame=True)
cancer
```

```python
df=cancer.frame
df
```

```python
df.shape
```

```python
print("original Data Frame",df.shape)
```

```python
X=df[cancer["feature_names"]]
X
```

```python
X_mean=X.mean()
X_std=X.std()
Z=(X-X_mean)/X_std
Z
```

```python
c=Z.cov()
c
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(c,annot=True)
plt.show()
```

```python
import numpy as np
eigenvalues,eigenvectors=np.linalg.eig(c)
```

```python
eigenvalues
```

```python
eigenvalues.shape
```

```python
eigenvectors
```

```python
eigenvectors.shape
```

```python
explained_var=np.cumsum(eigenvalues)/np.sum(eigenvalues)
explained_var
```

```python
n_components=np.argmax(explained_var>=0.50)+1
n_components
```
