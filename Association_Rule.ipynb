```python
# Convert string into list of items (not characters)
df["Items"] = df["Items"].apply(lambda x: [item.strip() for item in str(x).split(",")])

print(df.head(10))
```

```python
transaction=df["Items"].tolist()
transaction
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
te=TransactionEncoder()
te
```

```python
te_arry=te.fit_transform(transaction)
te_arry
```

```python
dataset=pd.DataFrame(te_arry,columns=te.columns_)
dataset 
```

```python
# def is a function,encode_unit is a function name
def encode_units(x):
    if x== False:
        return 0
    if x==True:
       return 1

# CATEGORICAL DATA conerted INTO BOOLEAN  that is 0 and 1
```

```python
dataset=dataset.applymap(encode_units)
dataset
```

```python
frequent_itemsets = apriori(dataset, min_support=.10,max_len=2, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=.20)
rules.head(20)
```

```python
#Last Step
frequent_itemsets = apriori(dataset, min_support=.01,max_len=2, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0)
rules

#frequent_itemsets : works on only numerical data 
# it will not work on categorical data 
```

```python
rules = rules.sort_values(["confidence", "lift"], ascending=[False, False])
rules
```
