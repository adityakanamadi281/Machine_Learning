```python
# Feature Engineering
x=df.drop("Price",axis=1)
x
```

```python
# converting Price , rupees to EUROs

conversion_rate = 0.011
rupees= df["Price"]
euros= rupees * conversion_rate
euros
```

```python
y=euros
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model1=LinearRegression()
model1.fit(x_test,y_test)
```

```python
y_pred=model.predict(x_train)
y_pred
```

```python
y_pred1=model1.predict(x_test)
y_pred1
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred1)
score
```

```python
score1=r2_score(y_train, y_pred)
score1
```

```python
from sklearn.linear_model import Lasso, Ridge
```

```python
lasso=Lasso(alpha=50)
```

```python
lasso.fit(x_train, y_train)
```

```python
lasso_pred=lasso.predict(x_test)
lasso_pred
```

```python
lasso_score=r2_score(y_test, lasso_pred)
lasso_score
```

```python
lasso_pred1=lasso.predict(x_train)
lasso_pred1
```

```python
lasso_score1=r2_score(y_train, lasso_pred1)
lasso_score1
```

```python
#Ridge technique
ridge=Ridge(alpha=10)
```

```python
ridge.fit(x_train,y_train)
```

```python
ridge_pred=ridge.predict(x_test)
ridge_pred
```

```python
ridge_score=r2_score(y_test, ridge_pred)
ridge_score
```
